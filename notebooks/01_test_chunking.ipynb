```python
%load_ext autoreload
%autoreload 2

import sys
from pathlib import Path

project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

from src.llm_platform.data_foundry.document_processor import DocumentProcessor

DATA_DIR = project_root / "data"
SAMPLE_PDF = DATA_DIR / "sample.pdf"

print(f"Project root: {project_root}")
print(f"Looking for PDF at: {SAMPLE_PDF}")
```

```python
processor = DocumentProcessor(chunk_size=800, chunk_overlap=150)

try:
    chunks = processor.process_file(SAMPLE_PDF)
    print(f"Successfully extracted {len(chunks)} chunks.")
except Exception as e:
    print(f"Error processing document: {e}")
```

```python
# Display the first 3 chunks to verify overlap and semantic boundaries
for i, chunk in enumerate(chunks[:3]):
    print("=" * 80)
    print(f"CHUNK {i + 1} | ID: {chunk.chunk_id}")
    print(f"Metadata: {chunk.metadata}")
    print("-" * 80)
    print(chunk.text)
    print("=" * 80)
    print("\n")
```
